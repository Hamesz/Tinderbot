```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import time
```

```python
# create 
landmark_attractive_df = pd.read_pickle("./Scripts/landmarks_attractive_dataframe.pkl")
```

```python
landmark_attractive_df.head()
```

```python
# create dataframe with column per point + label
landmark_attractive_df.iloc[0][-1]
```

```python
def flatten_tuple_list(t):
    flat_list = []
#     for sublist in t:
#         for item in sublist:
#             flat_list.append(item)
    flat_list = [x for sublist in t for x in sublist]
    return flat_list

columns = ['column_{}'.format(i) for i in range(144)]
columns.append('label')
new_df = pd.DataFrame(columns=columns)

start = time.time()
for index, row in landmark_attractive_df.iterrows():
    flattened_columns = landmark_attractive_df.iloc[index][:-1].apply(flatten_tuple_list)
    flattened_row = flatten_tuple_list(flattened_columns.values)
    flattened_row.append(row[-1])
    new_df.loc[index] = flattened_row
    if index % 2000 == 0:
        taken = time.time() - start
        start = time.time()
        print(f'{index} done in {taken} s!')
#     if index > 10000:
#         break
#     print(len(flattened_row))
```
