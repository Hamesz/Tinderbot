```python
import tensorflow as tf
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
landmark_points_csv = r"C:\Users\James H\git\Tinderbot\data\text\landmarks_points.csv"
```

```python
df = pd.read_csv(landmark_points_csv)
df['label'] = df['label'].apply(lambda x: 0 if x<0 else 1)
X = df.drop(columns='label')
Y = df['label']

# X_train = X.iloc[:8500]
# X_test = X.iloc[8500:]

# Y_train = Y.iloc[:8500]
# Y_test = Y.iloc[8500:]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.15)
# X_train, X_val, y_train, y_val  = train_test_split(X_train, y_train, test_size=0.15)
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.Input(shape=(144,)))
model.add(tf.keras.layers.Dense(72, activation='relu'))
model.add(tf.keras.layers.Dense(72, activation='relu'))
model.add(tf.keras.layers.Dense(64, activation='relu'))
model.add(tf.keras.layers.Dense(32, activation='relu'))
# model.add(tf.keras.layers.Dropout(0.2))

# model.add(tf.keras.layers.Dense(20, activation='relu'))
# model.add(tf.keras.layers.Dense(20, activation='relu', kernel_constraint=tf.keras.constraints.MaxNorm(3)))
# model.add(tf.keras.layers.Dense(10, activation='relu', kernel_constraint=tf.keras.constraints.MaxNorm(3)))
model.add(tf.keras.layers.Dense(1))
```

```python
base_learn_rate = 0.00001
model.compile(optimizer=tf.keras.optimizers.Adam(lr=base_learn_rate),
              loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              metrics=['accuracy'])

# sgd = tf.keras.optimizers.SGD(lr=base_learn_rate, momentum=0.8)
# model.compile(optimizer=sgd,
#               loss='binary_crossentropy',
#               metrics=['accuracy'])
```

```python
batch_size = 124
initial_epochs = 200
model.evaluate(x=X, y=Y, batch_size=batch_size, verbose=1)
```

```python
model.fit(x=X_train,
          y=y_train,
          batch_size=batch_size,
          epochs=initial_epochs,
          verbose=1,
          validation_split=0.15,
          shuffle=True,
          use_multiprocessing=True,
          workers=100
         )
```

```python
model.evaluate(x=X_test, y=y_test, batch_size=batch_size, verbose=1)
```
